```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# extracting courses available on the website 

base_url = 'https://talentedge.com/browse-courses'

driver = webdriver.Chrome()
# driver.maximize_window()
driver.get(base_url)
    
html_content = driver.page_source
soup = BeautifulSoup(html_content, 'html5lib')

course_link = []
data = soup.find_all('div', attrs = {'class':'know-more-p'})
for dt in data:
    temp_dt = dt.find('a', attrs = {'class':'view-all'})
    link = temp_dt['href']
    course_link.append(link)

driver.close()
# course_link
```

```python
# scraping all the required data

course_title = []
course_description = []
course_duration = []
course_start = []
course_wywl = []
course_skills = []
target_std = []
course_eligibility = []
course_content = []
faculty_name = []
faculty_designation = []
faculty_description = []
institute_name = []
fees_inr = []
fees_usd = []


for url in course_link:
    driver = webdriver.Chrome()
    # driver.maximize_window()
    driver.get(url)
    
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html5lib')
    
# finding course title
    try:
        if soup.find('h1', attrs = {'class':'pl-title'}):
            data = soup.find('h1', attrs = {'class':'pl-title'})
            title = data.get_text()
        else:
            temp_data = soup.find('div', attrs = {'class':'el-ln-banner-left'})
            data = soup.find('h1')
            title = data.get_text()
            
        course_title.append(title)
        
    except Exception as e:
        course_title.append(np.nan)
        print('Error finding title:', e)


# finding course description
    try:
        if soup.find('div', attrs = {'class':'desc_less'}):
            temp_data = soup.find('div', attrs = {'class':'desc_less'})
            data = temp_data.find('p')
            des = data.get_text()
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section2-inner el-ln-p'})
            data = temp_data.find_all('p')
            des = []
            for dt in data:
                des.append(dt.get_text())
            des = ' '.join(des)
            
        course_description.append(des)
        
    except Exception as e:
        course_description.append(np.nan)
        print('Error finding description:', e)


# finding course duration 
    try:
        if soup.find('div', attrs = {'class':'duration-of-course'}):
            temp_data = soup.find('div', attrs = {'class':'duration-of-course'})
            data = temp_data.find('strong')
            duration = data.get_text()
        else:
            data = soup.find('div', attrs = {'class':'months pb-1 pt-2'})
            duration = data.get_text()
            
        course_duration.append(duration)
    
    except Exception as e:
        course_duration.append(np.nan)
        print('Error finding duration:', e)


# finding course start time
    try:
        if soup.find('div', attrs = {'class':'duration-of-course'}):
            temp_data = soup.find('div', attrs = {'class':'duration-of-course'})
            data = temp_data.find_next('strong').find_next('strong')
            start = data.get_text()
            
        else:
            data = soup.find('h2', attrs = {'class':'no'})
            start = data.get_text()
            
        course_start.append(start)
        
    except Exception as e:
        course_start.append(np.nan)
        print('Error finding start time:', e)


# finding course what you will do
    try:
        if soup.find('div', attrs = {'class':'pl-deeper-undstnd to_flex_ul'}):
            temp_data = soup.find('div', attrs = {'class':'pl-deeper-undstnd to_flex_ul'})
            data = temp_data.find_all('li')
            
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section3-inner'})
            data = temp_data.find_all('strong')

        wywl = []
        for dt in data:
            wywl.append(dt.get_text())
        wywl = ' | '.join(wywl)
            
        course_wywl.append(wywl)
    
    except Exception as e:
        course_wywl.append(np.nan)
        print('Error finding what you will learn:', e)


# finding course skills
    try:
        temp_data = soup.find('div', attrs = {'class':'key-skills-sec'})
        data = temp_data.find_all('li')
        skills = []
        
        for dt in data:
            skills.append(dt.get_text())
        skills = ' | '.join(skills)
        
        course_skills.append(skills)
    
    except Exception as e:
        course_skills.append(np.nan)
        print('Error finding course skills:', e)


# finding targeted students by course
    try:
        if soup.find('h4', attrs = {'class':'cs-titlec'}):
            data = soup.find('h4', attrs = {'class':'cs-titlec'})
            target = data.get_text()
        else:
            temp_data = soup.find('div', attrs = {'class':'tab-content current'})
            data = temp_data.find_all('li')
            target = []
            for dt in data:
                target.append(dt.get_text())
            target = ' | '.join(target)

        target_std.append(target)
    
    except Exception as e:
        target_std.append(np.nan)
        print('Error finding target students:', e)


# finding course eligibility criteria
    try:
        if soup.find('div', attrs = {'class':'eligible-right-top-list'}):
            temp_data = soup.find('div', attrs = {'class':'eligible-right-top-list'})
            
            if temp_data.find('li'):
                data = temp_data.find_all('li')
                
            else:
                data = temp_data.find('p')
                eligibility = data.get_text()

        else:
            temp_data = soup.find('div', attrs = {'class':'tab-content d-block'})
            data = temp_data.find_all('li')
            
        eligibility = []
        for dt in data:
            eligibility.append(dt.get_text())
        eligibility = ' | '.join(eligibility)
        
        course_eligibility.append(eligibility)

    except Exception as e:
        course_eligibility.append(np.nan)
        print('Error finding course eligibility:', e)
                    

# finding course content
    try:
        if soup.find('div', attrs = {'class':'sylab-tab-ul'}):
            temp_data = soup.find('div', attrs = {'class':'sylab-tab-ul'})
            data = temp_data.find_all('li')
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section4-inner'})
            data = temp_data.find_all('div', attrs = {'class':'accordionButton'})
        
        content = []
        for dt in data:
            content.append(dt.get_text())
        content = ' | '.join(content)
        
        course_content.append(content)
    
    except Exception as e:
        course_content.append(np.nan)
        print('Error finding course content:', e)
    

# finding institute name
    try:
        if soup.find('div', attrs = {'class':'plc-institute'}):
            temp_data = soup.find('div', attrs = {'class':'plc-institute'})
            data = temp_data.find('h4', attrs = {'class':'about-ititle'})
            
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section6-inner'})
            data = temp_data.find('div', attrs = {'class':'el-lap-sec-hd white'})
            
        institute = data.get_text()
        institute_name.append(institute)
        
    except Exception as e:
        institute_name.append(np.nan)
        print('Error finding institute name:', e)


# finding fees in inr
    try:
        if soup.find('div', attrs = {'class':'program-details-total-pay-amt d-flex align-items-center justify-content-between ruppes'}):
            temp_data = soup.find('div', attrs = {'class':'program-details-total-pay-amt d-flex align-items-center justify-content-between ruppes'})
            data = temp_data.find('div', attrs = {'class':'program-details-total-pay-amt-right'})
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section-fee'})
            data = temp_data.find('td', attrs = {'class':'programFeeTotalPrice'})
        
        fees = data.get_text()
        fees_inr.append(fees)
    
    except Exception as e:
        fees_inr.append(np.nan)
        print('Error finding fees in inr:', e)
        print(url)


# finding fees in usd
    try:
        temp_data = soup.find('div', attrs = {'class':'program-details-total-pay-amt d-flex align-items-center justify-content-between dolor'})
        data = temp_data.find('div', attrs = {'class':'program-details-total-pay-amt-right'})
        
        fees = data.get_text()
        fees_usd.append(fees)
    
    except Exception as e:
        fees_usd.append(np.nan)
        print('Error finding fees in usd:', e)


# finding faculty name
    try:
        if soup.find('div', attrs = {'class':'best-fdetail'}):
            temp_data = soup.find_all('div', attrs = {'class':'best-fdetail'})
            name = []
            
            for data in temp_data:                
                temp_dt = data.find('h4', attrs = {'class':'best-fname'})
                name.append(temp_dt.get_text())
            
            name = ' | '.join(name)
            
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section7-inner'})
            data = temp_data.find_all('h4')
            name = []
            for dt in data:
                name.append(dt.get_text())
            name = ' | '.join(name)

        faculty_name.append(name)
            
    except Exception as e:
        faculty_name.append(np.nan)
        print('Error finding faculty name', e)

    
# finding faculty designation
    try:
        if soup.find('div', attrs = {'class':'best-fdetail'}):
            temp_data = soup.find_all('div', attrs = {'class':'best-fdetail'})
            designation = []
            
            for data in temp_data:
                temp_dt = data.find('p', attrs = {'class':'best-fdesingnation'})
                designation.append(temp_dt.get_text())

            designation = ' | '.join(designation)
            
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section7-inner'})
            data = temp_data.find_all('p')
            designation = []
            for dt in data:
                designation.append(dt.get_text())
            designation = ' | '.join(designation)
        
        faculty_designation.append(designation)
            
    except Exception as e:
        faculty_designation.append(np.nan)
        print('Error finding faculty designation', e)

    
# finding faculty description
    try:
        if soup.find('p', attrs = {'id':'modal_faculty_description'}):
            data = soup.find_all('p', attrs = {'id':'modal_faculty_description'})
        
        else:
            temp_data = soup.find('div', attrs = {'class':'el-lap-section7-inner'})
            data = temp_data.find_all('div', attrs = {'class':'el-rd readMoreHide'})

        fdes = []
        for dt in data:
            fdes.append(dt.get_text())
        fdes = ' | '.join(fdes)
            
        faculty_description.append(fdes)
    
    except Exception as e:
        faculty_description.append(np.nan)
        print('Error finding faculty description:', e)
            
    driver.close()
```

```python
# cleaning data that was scraped

all_data = []
length = len(course_link)
for i in range(length):


# cleaning course titles
    try:
        title = course_title[i]
        title = title.strip()
        course_title[i] = title
    
    except Exception as e:
        print('Error:', e)


# cleaning course description
    try:
        des = course_description[i]
        rep = ['\n', '\xa0','Read More']
        if isinstance(des, str):
            des = des.replace('•', '')
            for r in rep:
                des = des.replace(r, ' ')
            des = des.strip()
        course_description[i] = des
        
    except Exception as e:
        print('Error:', e)


# cleaning course duration
    try:
        data = course_duration[i]
        rep = ['Duration', '\n', ':']
        if isinstance(data, str):
            for r in rep:
                data = data.replace(r, '')
            data = data.split(' ')
            
            duration = []
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    duration.append(dt)
                
            duration = ' '.join(duration)
            course_duration[i] = duration
    
    except Exception as e:
        print('Error', e)


# cleaning course what you will learn
    try:
        data = course_wywl[i]
        if isinstance(data, str):
            data = data.split('|')
            wywl = []
            for dt in data:
                dt = dt.strip()
                wywl.append(dt)
            
            wywl = ' | '.join(wywl)
            course_wywl[i] = wywl
            
    except Exception as e:
        print('Error', e)


# cleaning course skills
    try:
        data = course_skills[i]
        if isinstance(data, str):
            data = data.split('|')
            skill = []
            for dt in data:
                dt = dt.strip()
                skill.append(dt)
                
            skill = ' | '.join(skill)
            course_skills[i] = skill
            
    except Exception as e:
        print('Error', e)


# cleaning target students
    try:
        data = target_std[i]
        if isinstance(data, str):
            data = data.replace('\n', '')
            data = data.strip()
            target_std[i] = data
    except Exception as e:
        print('Error:', e)


# cleaning course eligibility criteria
    try:
        data = course_eligibility[i]
        rep = ['\n', '\xa0', '•']
        if isinstance(data, str):
            for r in rep:
                data = data.replace(r, '')
            
            elig = []
            data = data.split(" | ")
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    elig.append(dt)
               
            elig = ' | '.join(elig)
            course_eligibility[i] = elig
    
    except Exception as e:
        print('Error:', e)


# cleaning course content
    try:
        data = course_content[i]
        if isinstance(data, str):
            data = data.replace('\n', '')
            data = data.split('|')
            content = []
            for dt in data:
                dt = dt.strip()
                content.append(dt)

            content = ' | '.join(content)
            course_content[i] = content

    except Exception as e:
        print('Error:', e)


# cleaning institute name
    try:
        data = institute_name[i]
        if isinstance(data, str):
            data = data.replace('ABOUT', '')
            data = data.strip()
            institute_name[i] = data
    
    except Exception as e:
        print(e)


# cleaning fees in inr
    try:
        data = fees_inr[i]
        if isinstance(data, str):
            data = data.replace('\n', '')
            data = data.split(' ')
            fees = []
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    fees.append(dt)
            
            if len(fees) > 2:
                fees = fees[:4]
                
            fees = ' '.join(fees)
            fees_inr[i] = fees
    except Exception as e:
        print('Error:', e)
                    

# cleaning fees in usd
    try:
        data = fees_usd[i]
        if isinstance(data,str):
            data = data.replace('\n', '')
            data = data.split(' ')
            fees = []
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    fees.append(dt)
            
            if len(fees) > 2:
                fees = fees[:2]
                
            fees = ' '.join(fees)
            fees_usd[i] = fees
    except Exception as e:
        print('Error:', e)        


# cleaning faculty names
    try:
        data = faculty_name[i]
        if isinstance(data, str):
            data = data.replace('\n','')
            data = data.split('|')
            name = []
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    name.append(dt)
            name = ' | '.join(name)
            
            faculty_name[i] = name
        
    except Exception as e:
        print('Error:', e)


# cleaning faculty designation
    try:
        data = faculty_designation[i]
        if isinstance(data, str):
            data = data.replace('\n','')
            data = data.split('|')
            desig = []
            for dt in data:
                dt = dt.strip()
                if dt != '':
                    desig.append(dt)
            desig = ' | '.join(desig)
            
            faculty_designation[i] = desig
        
    except Exception as e:
        print('Error:', e)

#  adding data to a list to create dataset later
    try:
        link = course_link[i]
        title = course_title[i]
        des = course_description[i]
        duration = course_duration[i]
        start = course_start[i]
        wywl = course_wywl[i]
        skills = course_skills[i]
        target = target_std[i]
        eligibility = course_eligibility[i]
        content = course_content[i]
        name = faculty_name[i]
        designation = faculty_designation[i]
        institute = institute_name[i]
        inr = fees_inr[i]
        usd = fees_usd[i]
        
        temp_list = [link, title, des, duration, start, wywl, skills, target, eligibility, content, name, designation, institute, inr, usd]
        
        all_data.append(temp_list)
        
    except Exception as e:
        print('Error', e)
        
        
```

```python
# creating dataset
cols = ['Course Link', 'Title', 'Description', 'Duration', 'Start Date',
        'What you will learn', 'Skills', 'Target Students', 'Eligibity Criteria',
        'Content', 'Faculty Name(s)', 'Faculty Designation(s)', 'Institute Name', 
        'Fees in INR', 'Fees in USD']
temp_df = pd.DataFrame(all_data, columns = cols)
temp_df
```
