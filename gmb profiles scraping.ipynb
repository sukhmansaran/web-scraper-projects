```python
all_escape_room_names = []
all_addresses = []
all_phones = []
all_escape_room_urls = []
all_operation_hours = []
all_reviews = []
all_social_profiles = []

for res in restaurants:
    driver = webdriver.Chrome()
#     driver.maximize_window()
    driver.get(res)
    
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html5lib')

# find title for web search
    try:
        data = soup.find('title')
        title = data.get_text()
        
    except:
        title = None
        
        
# find the escape room name
    try:
        data = soup.find('div', attrs = {'data-attrid':'title'}) or soup.find('h2', attrs={'data-attrid': 'title'})
        escape_room_name = data.get_text()
        all_escape_room_names.append(escape_room_name)
    
    except:
        all_escape_room_names.append(None)
        if escape_room_name is None:
            print(f'Escape Room Name not found for {title}.')
        
    
# find the address for escape room
    try:
        data = soup.find('span',attrs ={'class':'LrzXr'})
        address = data.get_text()
        all_addresses.append(address)
    
    except:
        all_addresses.append(None)
        if escape_room_name is None:
            print(f'Address not found for {title}.')
        else:
            print(f'Address not found for {escape_room_name}.')

    
# find the phone for escape room
    try:
        data = soup.find('span',attrs = {'class':'LrzXr zdqRlf kno-fv'})
        phone = data.get_text()
        all_phones.append(phone)
    
    except:
        all_phones.append(None)
        if escape_room_name is None:
            print(f'Phone number not found for {title}')
        else:
            print(f'Phone number not found for {escape_room_name}')

        
# find website url for escape room
    try:
        data = soup.find('a', attrs = {'class':'n1obkb'}, href=True) or soup.find('a', attrs = {'class':'ab_button'}, href=True)
        escape_room_url = data['href']
        all_escape_room_urls.append(escape_room_url)
    
    except:
        all_escape_room_urls.append(None)
        if escape_room_name is None:
            print(f'Excape Room URL not found for {title}')
        else:
            print(f'Excape Room URL not found for {escape_room_name}')
        
# find operation hours for escape room
    try:
        temp_data = soup.find('table', attrs = {'class':'WgFkxc'})
        data = soup.find('tr', attrs = {'class':'K7Ltle'})
        operation_hours = data.get_text()
        all_operation_hours.append(operation_hours)
    
    except :
        all_operation_hours.append(None)
        if escape_room_name is None:
            print(f'Operation Hours not found for {title}')
        else:
            print(f'Operation Hours not found for {escape_room_name}')
        

# find reviews of escape room
    try:
        data = soup.find('span', attrs = {'class':'Aq14fc'})
        rating = data.get_text()
        all_reviews.append(rating)

    except:
        all_reviews.append(None)
        if escape_room_name is None:
            print(f'Reviews not found for {title}')
        else:
            print(f'Reviews not found for {escape_room_name}')
        

# find social profiles for escape room
    try:
        data = soup.find_all('g-link', attrs = {'class':'fl w23JUc ap3N9d'})
        social_profiles = []
    
        for dt in data:
            temp_dt = dt.find('a')
            url = temp_dt['href']
            social_profiles.append(url)
        
        all_social_profiles.append(social_profiles)
    
    except:
        all_social_profiles.append(None)
        if escape_room_name is None:
            print(f'Escape Room URL not found for {title}')
        else:
            print(f'Escape Room URL not found for {escape_room_name}')
    
    driver.quit()
```

```python
# stroing the data in dictionary 
data = {
    'Escape Room Name': all_escape_room_names,
    'Address': all_addresses,
    'Phone Number': all_phones,
    'URL': all_escape_room_urls,
    'Operation Hours': all_operation_hours,
    'Reviews': all_reviews,
    'Social Profiles': all_social_profiles
}
```

```python
# saving the data into json file
json_data = json.dumps(data, indent=4)

with open('gmb_profiles.json', 'w') as outfile:
    outfile.write(json_data)
```
